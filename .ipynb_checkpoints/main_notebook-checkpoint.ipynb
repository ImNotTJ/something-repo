```python
MALE_NAMES = ["liam", "noah", "oliver", "theodore", "james", "henry", "mateo", "elijah", "lucas", "william", "benjamin", "levi", "ezra", "sebastian", "jack", "daniel", "samuel", "michael", "ethan", "asher", "john", "hudson", "luca", "leo", "elias", "owen", "alexander", "dylan", "santiago", "julian", "david", "joseph", "matthew", "luke", "jackson", "maverick", "miles", "wyatt", "thomas", "isaac", "jacob", "mason", "gabriel", "anthony", "carter", "logan", "aiden", "grayson", "caleb", "cooper"]
FEMALE_NAMES = ["olivia", "emma", "amelia", "carlotte", "mia", "sophia", "isabella", "evelyn", "ava", "sofia", "camila", "harper", "luna", "eleanor", "violet", "aurora", "elizabeth", "eliana", "hazel", "chloe", "ellie", "nora", "gianna", "lily", "emily", "aria", "scarlett", "penelope", "zoe", "ella", "avery", "avigail", "mila", "lucy", "isla", "ivy", "layla", "lainey", "nova", "grace", "willow", "riley", "emilia", "naomi", "elena", "madison", "vaneltina", "victoria", "stella", "delilah"]

MALE_HEIGHTS = {
    "5'0" : "0.08th",
    "5'1" : "0.24th",
    "5'2" : "0.65th",
    "5'3" : "2nd",
    "5'4" : "4th",
    "5'5" : "7th",
    "5'6" : "13th",
    "5'7" : "22nd",
    "5'8" : "33rd",
    "5'9" : "46th",
    "5'10" : "59th",
    "5'11" : "72nd",
    "6'0" : "82nd",
    "6'1" : "90th",
    "6'2" : "94th",
    "6'3" : "97th",
    "6'4" : "99th",
}
FEMALE_HEIGHTS = {
    "4'7" : "0.08th",
    "4'8" : "0.27th",
    "4'9" : "0.76th",
    "4'10" : "2nd",
    "4'11" : "4th",
    "4'12" : "6th",
    "5'0" : "9th",
    "5'1" : "16",
    "5'2" : "26",
    "5'3" : "39",
    "5'4" : "52",
    "5'5" : "67",
    "5'6" : "78",
    "5'7" : "87",
    "5'8" : "93",
    "5'9" : "97",
    "5'10" : "98",
    "5'11" : "99",
}

name = input ("Please give a name")

gender = -2
if name.lower() in MALE_NAMES:
    gender = 1
elif name.lower() in FEMALE_NAMES:
    gender = -1
else:
    while gender != -1 and gender != 0 and gender != 1:
        try:
            gender = int(input("Name not recognized! Please enter a 1 if you are male, or a 2 if you are female. This program does not work for any other possibilities, so in the case of being neither, type 0 to exit the program."))
        except ValueError:
            print("Please enter a valid number!")

if gender != 0:
    height_feet = -1
    while True:
        try:
            height_feet = int(input("Please give the feet-component of your height"))
            break
        except ValueError:
            print("Please enter a valid number!")
    height_inches = -1
    while True:
        try:
            height_inches = int(input("Please give the inch-component of your height"))
            break
        except ValueError:
            print("Please enter a valid number!")
    key = str(height_feet) + "'" + str(height_inches)

    percentile = ""
    dictionary = {}
    if gender == -1:
        dictionary = FEMALE_HEIGHTS
    elif gender == 1:
        dictionary = MALE_HEIGHTS

    if key in dictionary.keys():
        percentile = dictionary[key]
    else:
        if height_feet < 5:
            percentile = "0.08th"
        else:
            percentile = "99th"
print("You are in the " + percentile + " percentile for the US!")

age = input("Howie Mandel: How old are you?")

print(f"You: I'm {age} years old!\nHowie Mandel: You are the youngest person ever!")
```
